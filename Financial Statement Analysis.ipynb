## **Financial Performance Dashboard Overview**

### **1. Purpose**  
- Interactive dashboard comparing **current vs previous year financial ratios**.  

### **2. Data Processing**  
- Calculates **percentage changes** for key metrics.  

### **3. Dashboard Layout**  
- **Tabs** for different financial ratio categories.  
- **Bar charts** for visual comparison.  
- **Table** displaying numerical values with trend indicators.  

### **4. Technology Used**  
- Dash, Plotly, Pandas, Bootstrap.  

### **5. Execution**  
- Runs as a **web app** on port **8050** for real-time financial analysis.  


```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
from dash import Dash, dcc, html
import dash_bootstrap_components as dbc
# In this case, current year metrics can be a dictionary and py metrics can also be considered as on e. the prpare data command calculates % changes and organizes all neatly.The dash app creates a dash application using bootstap for styling and crate layout os used for creating the structure of the dashboard
class FinancialPerformanceDashboard:
    def __init__(self, current_year_metrics, previous_year_metrics):
        self.current_year = current_year_metrics
        self.previous_year = previous_year_metrics
        self.prepare_data()
        self.app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
        self.create_layout()
    
    def prepare_data(self):
        metric_categories = {
            'Profitability Ratios': [
                'Gross Profit Margin', 
                'Net Profit Margin', 
                'Profit Before Tax Margin', 
                'Return on Equity (ROE)', 
                'Return on Assets (ROA)'
            ],
            'Liquidity Ratios': [
                'Current Ratio', 
                'Quick Ratio'
            ],
            'Efficiency Ratios': [
                'Inventory Turnover', 
                'Days Inventory on Hand'
            ],
            'Leverage Ratios': [
                'Debt-to-Equity Ratio', 
                'Debt-to-Assets Ratio'
            ],
            'Shareholder Ratios': [
                'Earnings Per Share (EPS)', 
                'Dividend Per Share (DPS)', 
                'Dividend Payout Ratio', 
                'Earnings Retention Rate', 
                'Dividend Yield', 
                'Price-to-Earnings (P/E) Ratio'
            ]
        }
        
        
        data = []
        for category, metrics in metric_categories.items():
            for metric in metrics:
                current_value = self.current_year.get(metric, 0)
                previous_value = self.previous_year.get(metric, 0)
                
    
                if isinstance(current_value, str) and '%' in current_value:
                    current_value = float(current_value.replace('%', ''))
                if isinstance(previous_value, str) and '%' in previous_value:
                    previous_value = float(previous_value.replace('%', ''))
                
                
                percent_change = ((current_value - previous_value) / previous_value * 100) if previous_value != 0 else 0
                data.append({
                    'Category': category,
                    'Metric': metric,
                    'Current Year': current_value,
                    'Previous Year': previous_value,
                    'Percent Change': round(percent_change, 2)
                })
        
        self.comparative_df = pd.DataFrame(data)
    
    def create_layout(self):
        self.app.layout = dbc.Container([
            html.H1('Comprehensive Financial Performance Dashboard', className='txt'),
          
            dbc.Tabs([
               
                dbc.Tab(label='Profitability Ratios', children=[
                    dbc.Row([
                        dbc.Col([
                            dcc.Graph(id='profitability-comparison', 
                                      figure=self.create_category_comparison_chart('Profitability Ratios'))
                        ], width=12)
                    ])
                ]),
                
                
                dbc.Tab(label='Liquidity Ratios', children=[
                    dbc.Row([
                        dbc.Col([
                            dcc.Graph(id='liquidity-comparison', 
                                      figure=self.create_category_comparison_chart('Liquidity Ratios'))
                        ], width=12)
                    ])
                ]),
                
                
                dbc.Tab(label='Efficiency Ratios', children=[
                    dbc.Row([
                        dbc.Col([
                            dcc.Graph(id='efficiency-comparison', 
                                      figure=self.create_category_comparison_chart('Efficiency Ratios'))
                        ], width=12)
                    ])
                ]),
                
                
                dbc.Tab(label='Leverage Ratios', children=[
                    dbc.Row([
                        dbc.Col([
                            dcc.Graph(id='leverage-comparison', 
                                      figure=self.create_category_comparison_chart('Leverage Ratios'))
                        ], width=12)
                    ])
                ]),
                
                
                dbc.Tab(label='Shareholder Ratios', children=[
                    dbc.Row([
                        dbc.Col([
                            dcc.Graph(id='shareholder-comparison', 
                                      figure=self.create_category_comparison_chart('Shareholder Ratios'))
                        ], width=12)
                    ])
                ]),
            ]),
            
            
            dbc.Row([
                dbc.Col([
                    dbc.Card([
                        dbc.CardHeader(html.H3('Detailed Financial Metrics')),
                        dbc.CardBody([
                            html.Table(
                                # Table Header
                                [html.Thead(html.Tr([
                                    html.Th('Category'),
                                    html.Th('Metric'),
                                    html.Th('Current Year'),
                                    html.Th('Previous Year'),
                                    html.Th('Change %')
                                ]))] +
                                
                                [html.Tbody([
                                    html.Tr([
                                        html.Td(row['Category']),
                                        html.Td(row['Metric']),
                                        html.Td(f"{row['Current Year']:.2f}"),
                                        html.Td(f"{row['Previous Year']:.2f}"),
                                        html.Td(
                                            f"{row['Percent Change']:.2f}%", 
                                            style={'color': 'green' if row['Percent Change'] >= 0 else 'red'}
                                        )
                                    ]) for _, row in self.comparative_df.iterrows()
                                ])
                            ], className='table table-striped table-hover')
                        ])
                    ])
                ], width=12)
            ])
        ], fluid=True)
    
    def create_category_comparison_chart(self, category):
        category_df = self.comparative_df[self.comparative_df['Category'] == category]
        
        fig = px.bar(
            category_df, 
            x='Metric', 
            y=['Current Year', 'Previous Year'],
            title=f'{category} Comparison',
            barmode='group',
            labels={'value': 'Value', 'variable': 'Year'}
        )
        
        
        fig.update_layout(
            xaxis_title='Metrics',
            yaxis_title='Value',
            legend_title='Year',
            height=500
        )
        
        return fig
    
    def run_dashboard(self, port=8050):
        """
        Run the dashboard
        
        :param port: Port number to run the dashboard
        """
        self.app.run(debug=True, port=port)


if __name__ == '__main__':
    current_year_metrics = {
        'Gross Profit Margin': 52.27,
        'Net Profit Margin': 16.46,
        'Profit Before Tax Margin': 22.26,
        'Return on Equity (ROE)': 19.84,
        'Return on Assets (ROA)': 13.12,
        
        'Current Ratio': 1.64,
        'Quick Ratio': 1.33,
        
        'Inventory Turnover': 7.69,
        'Days Inventory on Hand': 47.45,
        
        'Debt-to-Equity Ratio': 0.0,
        'Debt-to-Assets Ratio': 0.0,

        'Earnings Per Share (EPS)': 43.05,
        'Dividend Per Share (DPS)': 40,
        'Dividend Payout Ratio': 92.92,
        'Earnings Retention Rate': 7.08,
        'Dividend Yield': 1.8,
        'Price-to-Earnings (P/E) Ratio': 51.59
    }

    previous_year_metrics = {
        
        'Gross Profit Margin': 47.91,
        'Net Profit Margin': 16.66,
        'Profit Before Tax Margin': 21.88,
        'Return on Equity (ROE)': 19.84,
        'Return on Assets (ROA)': 13.87,

        'Current Ratio': 1.38,
        'Quick Ratio': 1.03,

        'Inventory Turnover': 7.73,
        'Days Inventory on Hand': 47.31,

        'Debt-to-Equity Ratio': 0.0,
        'Debt-to-Assets Ratio': 0.0,

        'Earnings Per Share (EPS)': 33.21,
        'Dividend Per Share (DPS)': 28.20,
        'Dividend Payout Ratio': 84.91,
        'Earnings Retention Rate': 15.09,
        'Dividend Yield': 1.28,
        'Price-to-Earnings (P/E) Ratio': 66.56
    }
    dashboard = FinancialPerformanceDashboard(current_year_metrics, previous_year_metrics)
    dashboard.run_dashboard()
```

```python
import pandas as pd
import plotly.express as px
import dash_bootstrap_components as dbc
from dash import Dash, dcc, html
import FinancialRatios as FR  # this is the main module that is to be imported

class FinancialPerformanceDashboard:
    def __init__(self, file_name):
        self.analyzer = FR.FinancialRatios(file_name) # it draws on the main fr codule
        self.results = self.analyzer.mainFunction()
        self.current_year = self.results["Current Year"]
        self.previous_year = self.results["Previous Year"]
        self.prepare_data()
        self.app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
        self.create_layout()

    def prepare_data(self):
        # Add exogenous data
        self.exogenous_data = {
            "CY Market Price": 2216.39,
            "CY Shares Outstanding": 2349606299,
            "PY Market Price": 2210.25,
            "PY Shares Outstanding": 2999999998
        }

        metric_categories = {
            'Profitability Ratios': [
                'Gross Profit Margin', 'Net Profit Margin', 'Profit Before Tax Margin',
                'Return on Equity (ROE)', 'Return on Assets (ROA)'
            ],
            'Liquidity Ratios': ['Current Ratio', 'Quick Ratio'],
            'Efficiency Ratios': ['Inventory Turnover', 'Days Inventory on Hand'],
            'Leverage Ratios': ['Debt-to-Equity Ratio', 'Debt-to-Assets Ratio'],
            'Shareholder Ratios': [
                'Earnings Per Share (EPS)', 'Dividend Per Share (DPS)', 'Dividend Payout Ratio',
                'Earnings Retention Rate', 'Dividend Yield', 'Price-to-Earnings (P/E) Ratio'
            ]
        }

        data = []
        for category, metrics in metric_categories.items():
            for metric in metrics:
                current_value = self.current_year.get(category, {}).get(metric, 0)
                previous_value = self.previous_year.get(category, {}).get(metric, 0)

                if isinstance(current_value, str) and '%' in current_value:
                    current_value = float(current_value.replace('%', ''))
                if isinstance(previous_value, str) and '%' in previous_value:
                    previous_value = float(previous_value.replace('%', ''))

                percent_change = ((current_value - previous_value) / previous_value * 100) if previous_value else 0

                data.append({
                    'Category': category,
                    'Metric': metric,
                    'Current Year': current_value,
                    'Previous Year': previous_value,
                    'Percent Change': round(percent_change, 2)
                })

        self.comparative_df = pd.DataFrame(data)

    def create_layout(self):
        self.app.layout = dbc.Container([
            html.H1('Comprehensive Financial Performance Dashboard', className='text-center my-4'),

            dbc.Tabs([
                dbc.Tab(label=category, children=[
                    dbc.Row([
                        dbc.Col([
                            dcc.Graph(figure=self.create_category_comparison_chart(category))
                        ], width=12)
                    ])
                ]) for category in self.comparative_df['Category'].unique()
            ]),

            dbc.Row([
                dbc.Col([
                    dbc.Card([
                        dbc.CardHeader(html.H3('Detailed Financial Metrics')),
                        dbc.CardBody([
                            html.Table(
                                [html.Thead(html.Tr([
                                    html.Th('Category'), html.Th('Metric'),
                                    html.Th('Current Year'), html.Th('Previous Year'), html.Th('Change %')
                                ]))] +
                                [html.Tbody([
                                    html.Tr([
                                        html.Td(row['Category']), html.Td(row['Metric']),
                                        html.Td(f"{row['Current Year']:.2f}"),
                                        html.Td(f"{row['Previous Year']:.2f}"),
                                        html.Td(
                                            f"{row['Percent Change']:.2f}%",
                                            style={'color': 'green' if row['Percent Change'] >= 0 else 'red'}
                                        )
                                    ]) for _, row in self.comparative_df.iterrows()
                                ])
                            ], className='table table-striped table-hover')
                        ])
                    ])
                ], width=12)
            ])
        ], fluid=True)

    def create_category_comparison_chart(self, category):
        category_df = self.comparative_df[self.comparative_df['Category'] == category]

        fig = px.bar(
            category_df, x='Metric', y=['Current Year', 'Previous Year'],
            title=f'{category} Comparison', barmode='group',
            labels={'value': 'Value', 'variable': 'Year'}
        )

        fig.update_layout(
            xaxis_title='Metrics', yaxis_title='Value', legend_title='Year', height=500
        )

        return fig

    def run_dashboard(self, port=8050):
        self.app.run(debug=True, port=port)

if __name__ == '__main__':
    file_name = "Hindustan Unilever.xlsx"
    dashboard = FinancialPerformanceDashboard(file_name)
    dashboard.run_dashboard()
 # this is the exogeneous data : CY Market Price = Rs.2,216.39,CY Shares outstanding = 2,349,606,299,PY Market Price = Rs.â‚¹2,210.25,PY No of shares outstanding = 2,999,999,998
```
